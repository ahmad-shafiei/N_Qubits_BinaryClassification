```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import numpy as np
import random
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 1. Generate Synthetic Dataset
# -----------------------------------------------------------

def generate_synthetic_data(N=200):
    """
    Generate N sample measurement vectors (each 16 numbers)
    and assign labels based on the rule:
       mean(vector) < 0.5   -> class 0
       mean(vector) >= 0.5  -> class 1
    """
    np.random.seed(42)
    X = []
    y = []

    for _ in range(N):
        # Simulate 4 qubits -> 16 measurement outputs
        sample = np.random.rand(16)  # values in [0, 1]
        label = 1 if sample.mean() >= 0.5 else 0

        X.append(sample)
        y.append(label)

    X = np.array(X, dtype=np.float32)
    y = np.array(y, dtype=np.float32)

    return X, y

# -----------------------------------------------------------
# 2. Save Dataset to TXT File
# -----------------------------------------------------------

def save_dataset_to_txt(X, y, filename="qubit_dataset.txt"):
    """
    Saves dataset to a text file for later use.
    Format:
    x1 x2 x3 ... x16 label
    """
    with open(filename, "w") as f:
        for i in range(len(X)):
            row = " ".join(map(str, X[i])) + " " + str(int(y[i]))
            f.write(row + "\n")

    print(f"Dataset saved to {filename}")

save_dataset_to_txt(X, y, filename="qubit_dataset.txt")

# -----------------------------------------------------------
# -----------------------------------------------------------
# 3. Load Dataset from TXT File
# -----------------------------------------------------------

def load_dataset_from_txt(filename="qubit_dataset.txt"):
    """
    Reads dataset in the same format as saved above.
    """
    X = []
    y = []

    with open(filename, "r") as f:
        for line in f:
            values = list(map(float, line.strip().split()))
            X.append(values[:16])         # first 16 = features
            y.append(values[16])          # last = label

    print(f"Dataset loaded from {filename}")
    return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32)

X, y = load_dataset_from_txt("qubit_dataset.txt")
X,y = generate_synthetic_data(N=200)

# -----------------------------------------------------------
# 4. Define PyTorch Dataset and DataLoader


# -----------------------------------------------------------
# 2. PyTorch Dataset Class
# -----------------------------------------------------------

class QubitDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


# -----------------------------------------------------------
# 3. Define the Neural Network (Simple MLP)
# -----------------------------------------------------------

class QubitClassifier(nn.Module):
    def __init__(self):
        super(QubitClassifier, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)
    
# -----------------------------------------------------------
# 4. Prepare Data
# -----------------------------------------------------------

# Generate synthetic dataset
X, y = generate_synthetic_data(200)

# Shuffle indices for train/val/test split
indices = list(range(200))
random.shuffle(indices)

train_idx = indices[:140]    # 70%
val_idx   = indices[140:170] # 15%
test_idx  = indices[170:]    # 15%

X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val     = X[val_idx], y[val_idx]
X_test, y_test   = X[test_idx], y[test_idx]

# Create datasets and loaders
train_ds = QubitDataset(X_train, y_train)
val_ds = QubitDataset(X_val, y_val)
test_ds = QubitDataset(X_test, y_test)

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=16)
test_loader = DataLoader(test_ds, batch_size=16)


# -----------------------------------------------------------
# -----------------------------------------------------------
# 5. Training Setup
# -----------------------------------------------------------

model = QubitClassifier()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

EPOCHS = 50

# For plotting
loss_list = []
train_acc_list = []
val_acc_list = []
# -----------------------------------------------------------
# 6. Training Loop
# -----------------------------------------------------------

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0
    correct = 0

    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()

        outputs = model(batch_x).squeeze()
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

        predictions = (outputs >= 0.5).float()
        correct += (predictions == batch_y).sum().item()

    train_acc = correct / len(train_ds)

    # Validation
    model.eval()
    val_correct = 0

    with torch.no_grad():
        for batch_x, batch_y in val_loader:
            outputs = model(batch_x).squeeze()
            predictions = (outputs >= 0.5).float()
            val_correct += (predictions == batch_y).sum().item()

    val_acc = val_correct / len(val_ds)

    # Save metrics
    loss_list.append(train_loss)
    train_acc_list.append(train_acc)
    val_acc_list.append(val_acc)
                        

    if epoch % 10 == 0:
        print(f"Epoch {epoch:02d} | Train Loss: {train_loss:.3f} | "
              f"Train Acc: {train_acc:.3f} | Val Acc: {val_acc:.3f}")


# -----------------------------------------------------------
# 7. Testing
# -----------------------------------------------------------

model.eval()
test_correct = 0

with torch.no_grad():
    for batch_x, batch_y in test_loader:
        outputs = model(batch_x).squeeze()
        predictions = (outputs >= 0.5).float()
        test_correct += (predictions == batch_y).sum().item()

test_acc = test_correct / len(test_ds)

print("\nFinal Test Accuracy:", round(test_acc, 3))

# -----------------------------------------------------------
# 8. Save Model
# -----------------------------------------------------------

torch.save(model.state_dict(), "qubit_classifier.pt")
print("Model saved as qubit_classifier.pt")

#-----------------------------------------------------------
# 9. Plot Training Curves
# -----------------------------------------------------------

plt.figure(figsize=(14,4))

# Loss curve
plt.subplot(1, 3, 1)
plt.plot(loss_list, label="Train Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.legend()

# Training accuracy
plt.subplot(1, 3, 2)
plt.plot(train_acc_list, label="Train Accuracy", color="green")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy")
plt.grid(True)
plt.legend()

# Validation accuracy
plt.subplot(1, 3, 3)
plt.plot(val_acc_list, label="Validation Accuracy", color="red")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Validation Accuracy")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
